```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
with open('Emisiones_CO2.csv', 'r', encoding='latin-1') as archivo: #Abro el archivo con encoding latin por las tildes, ANSI también sirve
    lineas = archivo.readlines() #Uso el método readlines para que genere una lista de lineas
for linea in lineas[1:]:    #Elimino el header seleccionando de la segunda línea en adelante
    columns = linea.strip().split('|') #Separo las columnas por el separador '|' 
    dicc_emisiones['cod_pais'].append(columns[0]) #Agrego manualmente cada columna a su clave correspondiente en el diccionario
    dicc_emisiones['nom_pais'].append(columns[1])
    dicc_emisiones['region'].append(columns[2])
    dicc_emisiones['anio'].append(columns[3])
    dicc_emisiones['co2'].append(columns[4])
    dicc_emisiones['co2_percapita'].append(columns[5])
```

```python
print(dicc_emisiones)
```

```python
with open('Emisiones_CO2.csv', 'r', encoding='latin-1') as archivo: #Abro el archivo con encoding latin por las tildes
    lineas = archivo.readlines() #Uso el método readlines para que genere una lista de lineas
for linea in lineas[1:]:    #Elimino el header seleccionando de la segunda línea en adelante
    columns = linea.strip().split('|') #Separo las columnas por el separador '|' 
    for index,key in enumerate(dicc_emisiones.keys()): #Itero sobre los indices y las claves del diccionario
        dicc_emisiones[key].append(columns[index]) #Y voy agregando cada columna correspondiente
```

```python
with open('Emisiones_CO2.csv', 'r', encoding='latin-1') as archivo: #Abro el archivo con encoding latin por las tildes
    next(archivo)   #Me salto el header y empiezo por la segunda línea en adelante
    for linea in archivo:    
        columns = linea.strip().split('|') #Separo las columnas por el separador '|' 
        for index,key in enumerate(dicc_emisiones.keys()): #Itero sobre los indices y las claves del diccionario
            dicc_emisiones[key].append(columns[index]) #Y voy agregando cada columna correspondiente
```
