```python
def NumeroBinario(numero):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    if (not type(numero) == int) or (numero < 0):
        return None
    
    resto = [numero%2] #Divido entre dos y guardo el resto en una lista
    cociente = numero//2 #Divido entre dos y el cociente es la parte entera
    while cociente > 0: #Sigo dividiendo el cociente resultado entre dos hasta que sea cero
        resto.append(cociente%2) #Voy guardando los restos en la lista
        cociente = cociente//2
    binario = int(''.join(map(str,resto[::-1]))) #Invierto la lista y la convierto a str y luego a int usando map
    return binario
```

```python
# Código resultado del pair programming con Julian Alonso

def fra_bin(fraccion):

    ent = int(fraccion)
    dec = fraccion - ent

    p_entera = bin(int(ent))[2:]

    parte_decimal = ""

    while dec > 0:
        dec *= 2
        bit = int(dec)
        parte_decimal += str(bit)
        dec -= bit

    return p_entera + '.' + parte_decimal
fraccion = 6.3125
resultado = fra_bin(fraccion)
print(resultado)
```

```python
def FraccionBinario(fraccion, k):
    '''Esta función transforma un número decimal en número binario.
    param: 
    fraccion: Número decimal a transformar
    k: número de dígitos o precisión del decimal''' 
    
    entero = int(fraccion)
    decimal = fraccion - entero
    
    #Transformo primero la parte entera de la fracción
    lista_resto = [] 
    while entero > 0: 
        lista_resto.append(entero % 2) 
        entero = entero//2
    binario_entero = ''.join(map(str,lista_resto[::-1]))

    #Ahora transformo la parte decimal de la fracción
    lista_enteros = []
    while k > 0:  
        decimal = decimal * 2
        ent = int(decimal)
        lista_enteros.append(ent)
        decimal = decimal - ent
        k -=1
    binario_decimal = ''.join(map(str,lista_enteros))

    return binario_entero+'.'+binario_decimal
```

```python
resultado = FraccionBinario(1/3, 4)
print(resultado)
print(type(resultado))
```
